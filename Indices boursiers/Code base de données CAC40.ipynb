```python
import pandas as pd

# Récupérer la liste des entreprises du CAC 40 depuis Wikipedia
url = "https://en.wikipedia.org/wiki/CAC_40"
df_cac40 = pd.read_html(url)[4]  # Le tableau du CAC 40 est le quatrième tableau (index 3)

# Afficher les premières lignes pour vérifier
df_cac40.head()
```

```python
def get_market_cap(symbol):
    try:
        stock = yf.Ticker(symbol)
        market_cap = stock.info['marketCap']
        return market_cap
    except KeyError:
        return None
```

```python
df_cac40['MarketCap'] = df_cac40['Ticker'].apply(get_market_cap)
```

```python
df_cac40_MC = df_cac40.sort_values(by = 'MarketCap', ascending = False)
```
