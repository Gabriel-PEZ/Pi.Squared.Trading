```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
# Récupérer la liste des entreprises du S&P 500 depuis Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
df_sp500 = pd.read_html(url)[0]

# Afficher les premières lignes
df_sp500.head()
```

```python
#Je récupère la market cap avec cette fonction
def get_market_cap(symbol):
    try:
        stock = yf.Ticker(symbol)
        market_cap = stock.info['marketCap']
        return market_cap
    except KeyError:
        return None
```

```python
#Je crée une nouvelle colonne en appliquant ma fonction pour chaque valeur de la colonne Symbol
df_sp500['MarketCap'] = df_sp500['Symbol'].apply(get_market_cap)
```

```python
#J'utilise .isna() avec .any(axis = 1) qui permet de voir s'il y au moins (.any()) un NaN par ligne (axis = 1)
lignes_market_cap_na = df_sp500[df_sp500.isna().any(axis = 1)]
lignes_market_cap_na
```

```python
#En consultant Yahoo Finance, je récupère les ticker des entreprises correspondantes pour trouver leur marketCap avec yfinance

df_sp500.loc[60, 'Symbol'] = 'BRK-B'
df_sp500.loc[75, 'Symbol'] = 'BF-B'

df_sp500.loc[60, 'MarketCap'] = yf.Ticker('BRK-B').info['marketCap']
df_sp500.loc[75, 'MarketCap'] = yf.Ticker('BF-B').info['marketCap']
```

```python
df_sp500_MC = df_sp500.sort_values(by = 'MarketCap', ascending = False)
```
