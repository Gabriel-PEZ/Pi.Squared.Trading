```python
import pandas as pd

# Récupérer la liste des entreprises du CAC 40 depuis Wikipedia
url = "https://en.wikipedia.org/wiki/DAX"
df_dax40 = pd.read_html(url)[4]  # Le tableau du CAC 40 est le quatrième tableau (index 3)

# Afficher les premières lignes pour vérifier
df_dax40.head()
```

```python
df_dax40 = df_dax40.drop('Logo', axis = 1)
df_dax40 = df_dax40.drop('Employees', axis = 1)
```

```python
df_dax40
```

```python
def get_market_cap(symbol):
    try:
        stock = yf.Ticker(symbol)
        market_cap = stock.info['marketCap']
        return market_cap
    except KeyError:
        return None
```

```python
df_dax40['MarketCap'] = df_dax40['Ticker'].apply(get_market_cap)
```

```python
df_dax40 = df_dax40.sort_values(by = 'MarketCap', ascending = False)
```
